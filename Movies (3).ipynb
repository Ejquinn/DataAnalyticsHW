```python
#calculate decade
df['decade'] = (df['year'] // 10) * 10
```

```python
#uses group by to calculate average runtime and rating for each decade 
#formatted by reseting the index and prints results
decade_stats = df.groupby('decade').agg({'runtime': 'mean', 'rating': 'mean'}).reset_index()
print(decade_stats)
```

```python
#counts total number of movies released each year 
movies_per_year = df['year'].value_counts().sort_index()
print(movies_per_year)
```

```python
#finds the director with the most highly rated movies
df['director'] = df['stars'].str.extract(r'Director:\s*([^|\n]+)Stars:')[0].str.strip()
top_director = df[df['rating'] >= 8.0].groupby('director').size().idxmax()
print(f"Best Director:{top_director}")
#ChatGPT to find out specifically what to extract 
```

```python
#aggregate total movies and revenue by year 
yearly_stats = df.groupby('year').agg({'movies': 'count', 'gross': 'sum'}).rename(columns= {'movies': 'total_movies'}).reset_index()
print(yearly_stats)
```

```python
#pivot table for directors and revenue over the years
director_revenue = df.pivot_table(index='director', columns='year' , values='gross', aggfunc='sum')
print(director_revenue)
```

```python
#identify director with highest revenue
top_revenue_director = director_revenue.sum(axis=1).idxmax()
top_revenue_director = top_revenue_director.replace("\n","").strip()
print(f"Director with highest total revenue:{top_revenue_director}")
```

```python
#classify movies based on rating
df['rating_category'] = df['rating'].apply(lambda x: 'Excellent' if x >= 8.0 else 'Good' if x >= 6.5 else 'Average' if x >= 5.0 else 'Poor')
df['rating_category']
```

```python
# Identify high revenue movies 1 are high revenue or reach the 75% quantile if not they get a 0 
threshold = df['gross'].quantile(0.75)
df['high_revenue'] = df['gross'].apply(lambda x: 1 if x >= threshold else 0)
df['high_revenue']
```

```python
df.head()
```

```python
#calculates the relative votes for each movie  per genre
df['relative_votes'] = df.groupby('genre')['votes'].transform(lambda x: x/ x.sum() * 100)
```

```python
df.head()
```

```python
#creates a function called "classify_revenue()" that catergorizes movies based on gross revenue
# movies that grossed below the 33% are FLops movies that were above the 75% are moderate hits and everything else is a blockbuster
def classify_revenue(gross):
    if gross < df['gross'].quantile(0.33):
        return "Flop"
    elif gross < df['gross'].quantile(0.75):
        return "Moderate_Hit"
    else:
        return "Blockbuster"
```

```python
#adds column revenue_category to dataset 
df['revenue_category'] = df ['gross'].apply(classify_revenue)
```
